```python
import numpy as np
import pandas as pd
```

```python
covid_cases = pd.read_csv('../data/covid/time_series_covid19_confirmed_global.csv')
covid_cases.rename(columns = {'Country/Region': 'Country'}, inplace = True )
covid_cases = covid_cases[covid_cases['Country'].isin(['Ireland', 'US', 'United Kingdom']) & covid_cases['Province/State'].isnull()]
covid_cases = covid_cases.drop(['Province/State', 'Lat', 'Long'], 1)
covid_cases = covid_cases.set_index('Country')
covid_cases
```

```python
df = covid_cases.transpose()
df
```

```python
df['Ireland']
```

```python
from datetime import datetime, timezone
import dateutil.parser

df['day'] = [dateutil.parser.parse(d, dayfirst=False) for d in df.index] 
df
```

```python
difference = df.diff(axis=0)
difference['day'] = [dateutil.parser.parse(d, dayfirst=False) for d in difference.index] 
difference.fillna(0, inplace=True)
difference.set_index('day', inplace=True)
difference
```
